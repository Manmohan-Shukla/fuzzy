## numpy learn

```python
import numpy as np

a = np.array([1, 2, 3,4])
b=np.array([[133,23,32,23],[23,3,42,1]])
print(a, "\n ",b,"\n",a+b,"\n",a*b)
```

```python
aj=np.ones((3,2))
bj=np.zeros((1,5))
cj=np.full((2,3),4)
rand= np.random.random((2,3))
range= np.arange(2,4)
identity=np.eye(4)
print(aj)
print(bj)
print(cj)
print(rand)
print(range)
print(identity)
```

```python
a=np.array([[2,3,4],
           [3,5,4]])
print("shape",a.shape)
print("dimension",a.ndim)
print("size",a.size)
print("dtype",a.dtype)
```
